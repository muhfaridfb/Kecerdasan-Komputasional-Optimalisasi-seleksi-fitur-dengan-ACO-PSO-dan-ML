# Feature Selection

```python
# Handling outliers by replacing them with the IQR
df_outlier_handled = df_clean.copy()
numeric_columns = df_outlier_handled.select_dtypes(include=[np.number]).columns.tolist()
numeric_columns.remove('num')

for col in numeric_columns:
    Q1 = df_outlier_handled[col].quantile(0.25)
    Q3 = df_outlier_handled[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_outlier_handled[col] = np.where(df_outlier_handled[col] < lower_bound, lower_bound,
                                       np.where(df_outlier_handled[col] > upper_bound, upper_bound, df_outlier_handled[col]))

fig, axes = plt.subplots(nrows=5, ncols=3, figsize=(15, 10))
axes = axes.flatten()

for i, column in enumerate(df_outlier_handled.columns):
    if column != 'num':
        sns.boxplot(df_outlier_handled[column], ax=axes[i])
        axes[i].set_title(f'{column}')
    else:
        axes[i].axis('off')

plt.tight_layout()
plt.show()
```

```python
# Feature selection with chi square
from sklearn.feature_selection import chi2
from sklearn.preprocessing import MinMaxScaler

X = df_outlier_handled.drop(['num'], axis=1)
y = df_outlier_handled['num']

scaler = MinMaxScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

chi_scores = chi2(X_scaled, y)
p_values = pd.Series(chi_scores[1], index=X.columns).sort_values()

print("Chi-Square p-values:")
print(p_values)

selected_features = p_values[p_values < 0.05].index.tolist()
X_selected = X[selected_features]

print("\nSelected features based on chi-square (p < 0.05):")
print(selected_features)

df_selected = df_outlier_handled[selected_features + ['num']]
df_selected
```

# Modelling

```python
# Pisahkan fitur dan label
X = df_selected.drop('num', axis=1)
y = df_selected['num']

# Split jadi data train dan test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Cek shape hasil split
print(X_train.shape, X_test.shape)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
import random

def aco_pso_feature_selection(X, y, n_iter=30, n_particles=20):
    features = X.columns.tolist()
    best_score = 0
    best_features = features

    for _ in range(n_iter):
        for _ in range(n_particles):
            # Random pilih jumlah fitur
            n_selected = random.randint(3, len(features))
            selected_features = random.sample(features, n_selected)

            # Model dan cross-validation score
            model = RandomForestClassifier(random_state=42)
            score = cross_val_score(model, X[selected_features], y, cv=5, scoring='accuracy').mean()

            if score > best_score:
                best_score = score
                best_features = selected_features

    return best_features, best_score

# Jalankan seleksi fitur
selected_features, best_cv_score = aco_pso_feature_selection(X_train, y_train)

# Tampilkan hasil
print("Selected Features by ACO-PSO:", selected_features)
print("Best CV Score (Random Forest proxy):", best_cv_score)
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Random Forest
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train[selected_features], y_train)
y_pred_rf = rf_model.predict(X_test[selected_features])

# Evaluasi
print("\nRandom Forest Evaluation:")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```

```python
from xgboost import XGBClassifier

# XGBoost
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb_model.fit(X_train[selected_features], y_train)
y_pred_xgb = xgb_model.predict(X_test[selected_features])

# Evaluasi
print("\nXGBoost Evaluation:")
print("Accuracy:", accuracy_score(y_test, y_pred_xgb))
print(classification_report(y_test, y_pred_xgb))
```

# Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV

# Parameter grid Random Forest
param_grid_rf = {
    'n_estimators': [50, 100, 150],
    'max_depth': [3, 5, 7, None],
    'min_samples_split': [2, 4, 6],
    'max_features': ['auto', 'sqrt', 'log2']
}

# GridSearchCV untuk Random Forest
grid_rf = GridSearchCV(
    RandomForestClassifier(random_state=42),
    param_grid=param_grid_rf,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)

# Fit ke data train
grid_rf.fit(X_train[selected_features], y_train)

# Tampilkan hasil
print("Best Random Forest Params:", grid_rf.best_params_)
print("Best RF CV Score:", grid_rf.best_score_)
```

```python
# Parameter grid XGBoost
param_grid_xgb = {
    'n_estimators': [50, 100, 150],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}

# GridSearchCV untuk XGBoost
grid_xgb = GridSearchCV(
    XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    param_grid=param_grid_xgb,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)

# Fit ke data train
grid_xgb.fit(X_train[selected_features], y_train)

# Tampilkan hasil
print("Best XGBoost Params:", grid_xgb.best_params_)
print("Best XGBoost CV Score:", grid_xgb.best_score_)
```

```python
from sklearn.metrics import classification_report, accuracy_score

# Model RF terbaik
best_rf = grid_rf.best_estimator_
y_pred_rf_best = best_rf.predict(X_test[selected_features])

print("\nBest Random Forest Evaluation:")
print("Accuracy:", accuracy_score(y_test, y_pred_rf_best))
print(classification_report(y_test, y_pred_rf_best))

# Model XGB terbaik
best_xgb = grid_xgb.best_estimator_
y_pred_xgb_best = best_xgb.predict(X_test[selected_features])

print("\nBest XGBoost Evaluation:")
print("Accuracy:", accuracy_score(y_test, y_pred_xgb_best))
print(classification_report(y_test, y_pred_xgb_best))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

# Model RF terbaik
best_rf = grid_rf.best_estimator_

# Cross-validation di training data
cv_score_rf = cross_val_score(best_rf, X_train[selected_features], y_train, cv=5, scoring='accuracy').mean()
print(f"\nCross-Validation Accuracy (RF): {cv_score_rf:.4f}")

# Prediksi di test data
y_pred_rf_best = best_rf.predict(X_test[selected_features])
print("\nBest Random Forest Evaluation on Test Data:")
print("Accuracy:", accuracy_score(y_test, y_pred_rf_best))
print(classification_report(y_test, y_pred_rf_best))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf_best))


# Model XGB terbaik
best_xgb = grid_xgb.best_estimator_

# Cross-validation di training data
cv_score_xgb = cross_val_score(best_xgb, X_train[selected_features], y_train, cv=5, scoring='accuracy').mean()
print(f"\nCross-Validation Accuracy (XGBoost): {cv_score_xgb:.4f}")

# Prediksi di test data
y_pred_xgb_best = best_xgb.predict(X_test[selected_features])
print("\nBest XGBoost Evaluation on Test Data:")
print("Accuracy:", accuracy_score(y_test, y_pred_xgb_best))
print(classification_report(y_test, y_pred_xgb_best))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_xgb_best))
```
